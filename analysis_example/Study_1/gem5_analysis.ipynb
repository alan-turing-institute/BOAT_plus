```python
import copy as copy
import pandas as pd
from pandas.plotting import table
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
_RESULT_COLS = ["area", "power", "cycle", "success"]
```

```python
gem5_df = pd.read_csv("results.csv")
```

```python
def describe_result(result_df):
    """
    """
    
    for result_label in _RESULT_COLS:
        if result_label != "success":
            df = result_df[result_label]

            fig = plt.figure(figsize=(18,6))

            ax1 = plt.subplot(121)
            df.plot.hist(ax=ax1)

            ax2 = plt.subplot(122)
            tbl = table(ax2, df.describe(), loc='center')
            tbl.auto_set_font_size(False)
            tbl.set_fontsize(11)

            plt.axis('off') 
```

```python
describe_result(gem5_df)
```

```python
def perform_analysis(result_df):
    corr = result_df.corr()

    var_cols = list(set(list(result_df)) - set(_RESULT_COLS))
    res_cols = list(set(list(result_df)) - set(var_cols)) # should be the same as _RESULT_COLS
    
    sns.heatmap(corr.drop(res_cols, axis=1).drop(var_cols, axis=0))
```

```python
perform_analysis(gem5_df)
```
